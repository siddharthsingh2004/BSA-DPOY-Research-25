```python
import pandas as pd
```

```python
offensive_csv = pd.read_csv('../2024_offensive_advanced_metrics.csv')
aggregated_data = pd.read_csv('data/matchups_2024_agg.csv')
```

```python
mapping_player_rating = offensive_csv[['PLAYER', 'OFFRTG']]
mapping_player_rating
```

```python
aggregated_data['PLAYER'] = aggregated_data['first_name'].str.strip() + ' ' + aggregated_data['family_name'].str.strip()
merged_data = aggregated_data.merge(mapping_player_rating, on='PLAYER', how='left')
merged_data
```

```python
merged_data.to_csv('matchups_2024_agg_updated.csv', index=False)
```
