```python
import pandas as pd
```

```python
df = pd.read_csv('2023-24 Advanced Stats.xlsx - Sheet1.csv')
df.head()
df.info()
```

```python
df_new_stats = pd.read_csv("newStats.csv")
df_new_stats.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
# Ensure Personal Fouls (PF) is available in the main dataset
df_filtered = df.merge(df_new_stats[['Player', 'PF']], on='Player', how='left')

# Normalize PF and assign a negative weight (penalty for excessive fouling)
df_filtered['PF'] = -scaler.fit_transform(df_filtered[['PF']])  # Standardize PF

df_filtered['STL_per_100'] = df_filtered['STL%'] * df_filtered['MP'] / 100
df_filtered['BLK_per_100'] = df_filtered['BLK%'] * df_filtered['MP'] / 100
```

```python
df_filtered.head()
```

```python
df_filtered = df_filtered[df_filtered['MP'] > 1500]
```

```python
defensive_features = ['STL%', 'BLK%', 'DRB%', 'DBPM', 'DWS', 'BPM']
```

```python
team_avg_defense = df_filtered.groupby('Team')[defensive_features].transform('mean')
df_new_try = df_filtered[defensive_features] - team_avg_defense
```

```python
from sklearn.decomposition import PCA
```

```python
X_scaled = scaler.fit_transform(df_filtered[defensive_features])

# Apply PCA
pca = PCA(n_components=1)  # Extract the primary defensive component
df_filtered['Defensive_Score'] = pca.fit_transform(X_scaled)

# Standardize the final defensive score
df_filtered['Defensive_Score'] = (df_filtered['Defensive_Score'] - df_filtered['Defensive_Score'].mean()) / df_filtered['Defensive_Score'].std()
```
